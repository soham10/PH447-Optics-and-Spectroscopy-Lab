##Theory:
Cryptography is the technique of encrypting messages or data so that the information becomes unreadable to unauthorized persons and can only be understood by the sender and the intended receiver using a secret key. In classical cryptography, the security of the key depends on complex mathematical algorithms or computational difficulty, but there is always a possibility that the key may be cracked. Quantum cryptography overcomes this limitation by using the principles of quantum physics to ensure secure communication. The experiment demonstrates an analogy of quantum encryption using the BB84 protocol, which combines the concept of one-time pad encryption with quantum key distribution. In the one-time pad method, a random binary sequence of 0s and 1s is used as a key to encrypt a message.

Quantum key distribution (QKD) allows two users to generate this secret key securely using the polarization states of light. Any attempt to intercept the transmitted information disturbs the quantum states and can therefore be detected, ensuring secure communication.

The BB84 protocol uses two polarization bases of light: the + basis (0° and 90°) and the x basis (-45° and 45°). These polarizations represent binary bits 0 and 1. The sender, Alice, sends photons with random polarization in a random basis, and the receiver, Bob, measures them using a randomly chosen basis. They then compare the bases over a public channel and keep only the results where the bases match, which form the secret key. If an eavesdropper tries to intercept the photons, incorrect basis choices introduce errors that allow Alice and Bob to detect the intrusion.

Encryption Key: 111100010110111100000101

Word: FCB = 010001100100001101000010

```python
def xor(key: str, word: str) -> str:
  result = ''.join('1' if key[i] != word[i] else '0' for i in range(len(key))) # bit by bit
  return result
```

```python
key = "111100010110111100000101"
word = "010001100100001101000010" #FCB
print("Encrypted Result:", xor(key, word))
```

```python
key = "111100010110111100000101"
word = "101101110010110001000111"
print("Decrypted Result:", xor(key, word))
```
